## 1. Draw a line in a daigram from position (0,0) to position (6,100)

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([0,6])
y = np.array([0,100])

plt.plot(x,y)
plt.show()
```

## Mark each point with a circle

```python
import matplotlib.pyplot as plt
import numpy as np

#x = np.array([0,6])
y = np.array([3,8,1,10])

plt.plot(y,  'o:r',ms = 20,mfc = "b")
plt.show() 
```

```python
import matplotlib.pyplot as plt
import numpy as np

#x = np.array([0,6])
y = np.array([1,4,6])
plt.plot(y,c = "r",linewidth = "20.9")
plt.show() 
```

## add lables to the x-and y-axis

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([0,6])
y = np.array([0,100])

plt.plot(x,y)
plt.xlabel("Average pulse",c="r")
plt.ylabel("calorie Burnage",c="r")
plt.title("Sports Watch Data",c="b",size = "20",loc = "left")
plt.grid(axis = "x",c = "g",linestyle="dashed")
plt.show()
```

## subplot

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([0,6,5,4])
y = np.array([0,4,8,3])
plt.subplot(1,1,1)

plt.plot(x,y)
plt.xlabel("Average pulse",c="r")
plt.ylabel("calorie Burnage",c="r")
plt.title("Sports Watch Data",c="b",size = "20",loc = "left")
plt.grid(axis = "x",c = "g",linestyle="dashed")
plt.show()
```

## scatter plot

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([0,4,6,7,8,2,5,4,3])
y = np.array([0,1,5,4,7,6,8,9,2])
colors = np.array([1,3,4,5,6,7,8,9,2])

#plt.(x,y)
plt.scatter(x,y,  c =colors, cmap = "viridis")
plt.colorbar()
plt.show()
```

## Bar graph

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([0,2,4,5,6])
y = np.array([0,1,6,5,8])

plt.bar(x,y)
plt.show()
```
